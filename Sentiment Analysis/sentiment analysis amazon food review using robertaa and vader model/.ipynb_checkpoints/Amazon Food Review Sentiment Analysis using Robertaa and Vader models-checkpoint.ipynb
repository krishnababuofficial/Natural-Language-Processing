## Import Packages

```python

def preprocess_text(example):
    doc = nlp(example)
    
    filtered_text = []
    
    for token in doc:
        if not token.is_stop and not token.is_punct and token.is_alpha:
            filtered_text.append(token.lemma_.lower())
        
    return " ".join(filtered_text)
```

```python
example = df['Text'][1]
print(example)
```

```python
preprocess_text(example)
```

```python
df['preprocessed_text'] = df['Text'].apply(preprocess_text)
```

```python
df['preprocessed_text'][0:5]
```

## Vader Sentiment Score

```python
example_1 = df['preprocessed_text'][1]
print(example_1)
```

```python
# Lets see the example(text) with Vader

sia = SentimentIntensityAnalyzer()

sia.polarity_scores(example_1)
```

- The sentiment analysis results indicate that the text has a moderate level of negativity, with a negative score of 12.9%.
  
- The majority of the text is classified as neutral, with a score of 76.2%.
  
- There is also a minor positive sentiment, accounting for 11% of the text.
  
- The compound score of -0.1027 reflects a slightly negative overall sentiment, indicating a subtle leaning towards negativity in the analyzed text.

## Roberta Pre-Trained Model

- [Twitter-roBERTa-base for Sentiment Analysis](https://huggingface.co/cardiffnlp/twitter-roberta-base-sentiment-latest)   <- -     Link for model
    - This is a RoBERTa-base model trained on ~124M tweets.
    - And finetuned for sentiment analysis with the TweetEval benchmark.

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = TFAutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
# Lets see the example(text) with Roberta Model

encoded_input = tokenizer(example_1,  return_tensors='tf')
output = model(encoded_input)
scores = output[0][0].numpy()
scores = softmax(scores)
 
score_dict = {
    'Roberta_Negative' : scores[0],        # Labels: 0 -> Negative; 
    'Roberta_Neutral' : scores[1],         # Labels: 1 -> Neutral;          
    'Roberta_Positive' : scores[2]         # Labels: 2 -> Positive.
}

score_dict
```

- The model assigns about 36.57% positivity, indicating a substantial portion of the analyzed content is perceived as positive.

- Approximately 31.57% is identified as neutral, suggesting a balanced mix of neutral sentiments in the content.

- About 31.87% of the content is classified as negative, indicating a significant portion of the analyzed content is perceived as negative.

```python
# Made a def function from above code 

def polarity_score_roberta(example):
    encoded_input = tokenizer(example,  return_tensors='tf')
    output = model(encoded_input)
    scores = output[0][0].numpy()
    scores = softmax(scores)
    
    score_dict = {
    'Roberta_Negative' : scores[0],
    'Roberta_Neutral' : scores[1],
    'Roberta_Positive' : scores[2]
    }

    return score_dict
```

```python
# Run Vader and Roberta on entier data set

result = {}

for i, rows in tqdm(df.iterrows(), total = len(df)):
    try:
        text = rows['preprocessed_text']
        id = rows['Id']
        vader_results = sia.polarity_scores(text)
        #rename vader score results
        vader_results_rename = {}

        for key, value in vader_results.items():
            vader_results_rename[f'vader_{key}'] = value

        roberta_results = polarity_score_roberta(text)
        both_results = {**vader_results_rename, **roberta_results}
        result[id] = both_results
    except RuntimeError:
        print(f'Broke for id {id}')    
```

```python
# Store in padas DataFrame

results_df = pd.DataFrame(result).T
results_df = results_df.reset_index().rename(columns={'index': 'Id'}) # index as id - To merege this on to the original df
results_df = results_df.merge(df, how = 'left')
results_df.head()
```

## Vader Plots Result

```python
# Lets see for Vader_Compound

plt.style.use('ggplot')

sns.barplot(data = results_df, x = 'Score', y = 'vader_compound', errorbar = 'sd')
plt.title('Compound Score Amazon Star Review')
plt.show();
```

1. #### Compound Scores:    
    - The compound scores provide insight into how comments or reviews span the spectrum from negative to positive sentiment.   
    
2. #### Higher Star Ratings, Higher Compound Scores:
    - Reviews with 3, 4, and 5 stars show higher compound scores, suggesting a more positive sentiment.   
  
3. #### Lower Star Ratings, Lower Compound Scores:
    - Reviews with 1 and 2 stars have lower compound scores, indicating a more negative sentiment, especially with a high negative component.  

```python
# Vader_pos, Vader_neg, Vader_neu

fig, axs = plt.subplots(1, 3, figsize =(22,6))
sns.barplot(ax = axs[0], data = results_df, x = 'Score', y = 'vader_pos', errorbar = 'sd')
sns.barplot(ax = axs[1], data = results_df, x = 'Score', y = 'vader_neu', errorbar = 'sd')
sns.barplot(ax = axs[2], data = results_df, x = 'Score', y = 'vader_neg', errorbar = 'sd')
axs[0].set_title('Positive')
axs[1].set_title('Neutral')
axs[2].set_title('Negative')
plt.show();
```

1. #### Positive Sentiment:
    - Higher star ratings align with higher positivity scores, indicating a positive sentiment in reviews.

2. #### Neutral Sentiment:
    - Neutral sentiment appears consistent across various star ratings, presenting a relatively flat distribution.

3. #### Negativity Decreases with Higher Stars, Increases with Lower Stars:
    - As star ratings increase to 5, negativity tends to decrease, while lower star ratings correspond to higher levels of negativity in reviews.
